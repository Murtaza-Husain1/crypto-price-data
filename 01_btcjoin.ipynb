# BTCUSD Join to Altcoin Dataframe

> This module will connect BTCUSD daily data to the altcoin dataframes created in the first module. 

```python
#hide
from nbdev.showdoc import *
```

```python
import numpy as np
import pandas as pd
```

```python
btc_data = pd.read_csv('data/btc-usd-daily.csv')
```

```python
# a column with 'BTCUSD' for every element adds no information
btc_data = btc_data.drop(columns='Symbol')
```

```python
# rename 'Date' column to 'T' so that the two dataframes can be joined
btc_data.rename(columns={'Date':'T'}, inplace=True)
```

```python
btc_data[::-1].head()
```

```python
filename = 'btc-ltc-daily.csv'
```

```python
alt_data = pd.read_csv('data/' + filename)
alt_data = alt_data.drop(columns='Unnamed: 0')
```

```python
alt_data.head()
```

```python
joined_data = pd.merge(btc_data[::-1], alt_data, on='T')
```
